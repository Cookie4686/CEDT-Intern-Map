```python
import requests
import json
import pprint
import pandas as pd
import numpy as np
import time
import datetime
```

# Fetch All Available Application

```python
from google.colab import userdata
# Session determine the available application based on student's current year
session_id = userdata.get(' session_id')
# Athena Session Cookie from the internship web
cookie = userdata.get('cookie')
```

```python
search_query="onlyBookmarked=false&onlyAvailablePositions=true"

# Get total item count first
data_info = requests.get(
  url= f'https://cedtintern.cp.eng.chula.ac.th/api/sessions/{session_id}/openings?search=&page=1&limit=1&{search_query}',
  headers={
   "cookie": cookie
})
total_item_count = data_info.json()['meta']['totalItem']

# Get All Remaining Items
data_info = requests.get(
  url= f'https://cedtintern.cp.eng.chula.ac.th/api/sessions/{session_id}/openings?search=&page=1&limit={total_item_count}&{search_query}',
  headers={
   "cookie": cookie
})
data = data_info.json()
application_df = pd.DataFrame(data['items'])
application_df
```

```python
data_info = requests.get(
  url= f'https://cedtintern.cp.eng.chula.ac.th/api/common/opening-tags?sessionId={session_id}',
  headers={
   "cookie": cookie
})
tag_df = pd.DataFrame(data_info.json())
tag_df
```

```python
tag_df.to_json('tags.json', orient='records')
```

# Search The Address Location

```python
application_df['officeAddressLine1'] = application_df['officeAddressLine1'].apply(lambda x : " ".join(x.split()) )
```

```python
# Beware of google's bot detection on high usage
# When search in google map it will return coordinate to the web browser to shift the map there
# We locate that coordinate and store it

# Return ['longitude', 'latitude']
def search_location(address, verbose=False):
  address.replace('\n', ' ')
  if address and address != 'ไม่ระบุ':
    data_info = requests.get(
      url= f'https://www.google.com/search?tbm=map&hl=th&gl=th&q={address}',
    )
    text = data_info.text
    # text = text.replace('""','"')

    try:
      null = np.nan
      data = eval(text.removeprefix(")]}'"))
      res = data[1][0][1:]
      if verbose: print("Location Extract Result", res)
      return float(res[0]), float(res[1])
    except:
      print(text)
      return [np.nan, np.nan]

  return [np.nan, np.nan]
```

```python
address_group = application_df.groupby('officeAddressLine1')

address_df = pd.DataFrame(data=[],columns=['address', 'longitude', 'latitude'])
for address, _ in address_group:
  coords = search_location(address)
  row = pd.DataFrame(data=[[address, coords[0], coords[1]]], columns=['address', 'longitude', 'latitude'])
  address_df = pd.concat([address_df, row], axis=0, copy=False)

address_df
```

```python
for _, row in address_df.iterrows():
  filter_idx = application_df['officeAddressLine1'] == row['address']
  application_df.loc[filter_idx, ['latitude', 'longitude']] = [row['latitude'], row['longitude']]
application_df.sample(3)
```
